```python
import pandas as pd
```

Have weather document show 9am,12pm,3pm,5pm temperature times - turn into columuns 
Change column names

```python
weather['dt_iso'] = weather['dt_iso'].astype(str).str.replace(' +0000 UTC', '', regex=False)
weather['dt_iso'] = pd.to_datetime(weather['dt_iso'], format='%Y-%m-%d %H:%M:%S')
nine_data = weather[weather['dt_iso'].dt.hour == 9]
noon_data = weather[weather['dt_iso'].dt.hour == 12]
three_data = weather[weather['dt_iso'].dt.hour == 15]
five_data = weather[weather['dt_iso'].dt.hour == 17]
```

```python
nine_rename = {
    'temp':'Nine_AM_temp'
}

noon_rename = {
    'temp':'Noon_AM_temp'
}

three_rename = {
    'temp':'Three_PM_temp'
}

five_rename = {
    'temp':'Five_PM_temp'
}

nine_data.rename(columns=nine_rename, inplace=True)
noon_data.rename(columns=noon_rename, inplace=True)
three_data.rename(columns=three_rename, inplace=True)
five_data.rename(columns=five_rename, inplace=True)
```

```python
nine_needed_columns = ['dt_iso','Nine_AM_temp']
noon_needed_columns = ['dt_iso','Noon_AM_temp']
three_needed_columns = ['dt_iso','Three_PM_temp']
five_needed_columns = ['dt_iso','Five_PM_temp']
weather_needed_columns = ['dt_iso','weather_main']

nine_data_selected = nine_data[nine_needed_columns]
noon_data_selected = noon_data[noon_needed_columns]
three_data_selected = three_data[three_needed_columns]
five_data_selected = five_data[five_needed_columns]

#merge the 4 different hours on dt_iso

merged_hours = pd.merge(nine_data_selected, noon_data_selected, on='dt_iso', how='outer')
merged_hours1 = pd.merge(merged_hours, three_data_selected, on='dt_iso', how='outer')
merged_hours2 = pd.merge(merged_hours1, five_data_selected, on='dt_iso', how='outer')


```

```python
selected_columns = weather[['dt_iso','weather_main']]
weather_main_data = weather[weather['dt_iso'].dt.hour == 12]
new_weather_main = weather_main_data[weather_needed_columns]
final_merge = pd.merge(new_weather_main,merged_hours2,on='dt_iso',how='right')
```

```python
final_merge
```

merge to make one row for each date


```python
final_merge['dt_iso'] = pd.to_datetime(final_merge['dt_iso'])

final_merge['date'] = final_merge['dt_iso'].dt.date
final_merge['time'] = final_merge['dt_iso'].dt.time


result = final_merge.groupby('date').agg({
    'Nine_AM_temp':'first',
    'Noon_AM_temp':'first',
    'Three_PM_temp':'first',
    'Five_PM_temp':'first',
    'weather_main': 'first'
}).reset_index()
```

```python
storage['Booked Date'] = pd.to_datetime(storage['Booked Date'])
storage['date'] = storage['Booked Date'].dt.strftime('%Y-%m-%d')
```

```python
booking_count = storage.groupby('date').size().reset_index(name='Booked Boxes')
```

```python
booking_count
```

```python
date_range = pd.date_range(start='2023-01-01',end='2023-12-31',freq='D')
```

```python
storage_date_range = pd.DataFrame({'date':date_range})
```

```python
storage_date_range['date'] = storage_date_range['date'].astype(str)
```

```python
final_storage = pd.merge(storage_date_range,booking_count,on='date',how='left').fillna({'Booked Boxes': 0})
```

```python
combined_table = pd.merge(result,final_storage,on='date',how='left')
```
