Adding 9am, 3pm & 5pm Temps to get a more accurate view of daily temperature

```python
allhours = pd.read_excel('DallasWeatherAllHours.xlsx')
```

```python
allhours
```

place columns values into row - create columns (9am temp) (noon temp) (three temp (five temp))

```python
allhours['dt_iso'] = allhours['dt_iso'].astype(str).str.replace(' +0000 UTC', '', regex=False)
allhours['dt_iso'] = pd.to_datetime(allhours['dt_iso'], format='%Y-%m-%d %H:%M:%S')
nine_data = allhours[allhours['dt_iso'].dt.hour == 9]
three_data = allhours[allhours['dt_iso'].dt.hour == 15]
five_data = allhours[allhours['dt_iso'].dt.hour == 17]
```

```python
nine_rename = {
    'temp':'Nine_AM_temp'
}

three_rename = {
    'temp':'Three_PM_temp'
}

five_rename = {
    'temp':'Five_PM_temp'
}

nine_data.rename(columns=nine_rename, inplace=True)
three_data.rename(columns=three_rename, inplace=True)
five_data.rename(columns=five_rename, inplace=True)
```

```python
nine_needed_columns = ['dt_iso','Nine_AM_temp']
three_needed_columns = ['dt_iso','Three_PM_temp']
five_needed_columns = ['dt_iso','Five_PM_temp']

nine_data_selected = nine_data[nine_needed_columns]
three_data_selected = three_data[three_needed_columns]
five_data_selected = five_data[five_needed_columns]

#merge the 3 different hours on dt_iso

merged_hours = pd.merge(nine_data_selected, three_data_selected, on='dt_iso', how='outer')
merged_hours = pd.merge(merged_hours, five_data_selected, on='dt_iso', how='outer')
```

```python
merged_hours
```

```python
merged_hours['dt_iso'] = pd.to_datetime(merged_hours['dt_iso'])

merged_hours['date'] = merged_hours['dt_iso'].dt.date
merged_hours['time'] = merged_hours['dt_iso'].dt.time

result = merged_hours.groupby('date').agg({
    'Nine_AM_temp':'first',
    'Three_PM_temp':'first',
    'Five_PM_temp':'first'
}).reset_index()

```

```python
result
```

```python
result.to_excel('DallasWeatherAllHours.xlsx', index=False)
```
